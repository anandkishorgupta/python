```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data into a Pandas DataFrame
df = pd.read_csv('sales.csv')

# Remove dollar signs and convert columns to numeric values
df['Advertising Spend'] = df['Advertising Spend'].replace('[\$,]', '', regex=True).astype(float)
df['Sales Revenue'] = df['Sales Revenue'].replace('[\$,]', '', regex=True).astype(float)

# Calculate the correlation between advertising spending and sales revenue
correlation = df['Advertising Spend'].corr(df['Sales Revenue'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Advertising Spend', y='Sales Revenue', hue='Product Category', style='Sales Channel', palette='Set2')
plt.title(f'Scatter Plot - Advertising Spend vs. Sales Revenue\nCorrelation: {correlation:.2f}')
plt.xlabel('Advertising Spend')
plt.ylabel('Sales Revenue')
plt.legend(title='Category & Channel')
plt.show()
```
