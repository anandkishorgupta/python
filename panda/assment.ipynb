```python
import matplotlib.pyplot as plt
import numpy as np

# Sales revenue data
categories = ['Appliances', 'Clothing', 'Electronics']
Q1 = [165000, 525000, 630000]
Q2 = [105000, 420000, 460000]
percentage_change = [-36, -20, -27]

# Convert categories to numerical positions
x_positions = np.arange(len(categories))

# Calculate total revenue for each quarter
total_revenue_q1 = sum(Q1)
total_revenue_q2 = sum(Q2)

# Plotting the bar chart
fig, ax = plt.subplots()
bar_width = 0.35

# Plot bars for Q1
bars_q1 = ax.bar(x_positions, Q1, bar_width, label='Q1')

# Plot bars for Q2 with adjusted positions
bars_q2 = ax.bar(x_positions + bar_width, Q2, bar_width, label='Q2')

# Display percentage change as text on top of the bars
for i, value in enumerate(percentage_change):
    ax.text(x_positions[i] + bar_width/2, max(Q1[i], Q2[i]) + 20000, f'{value}%', ha='center')

# Add labels, title, and legend
ax.set_xlabel('Category')
ax.set_ylabel('Sales Revenue')
ax.set_title('Quarterly Sales Revenue by Category')
ax.set_xticks(x_positions + bar_width/2)
ax.set_xticklabels(categories)
ax.legend()

# Display total revenue for each quarter at the top of the chart
ax.text(1.5 * bar_width, max(total_revenue_q1, total_revenue_q2) + 50000,
        f'Total Q1: ${total_revenue_q1:,}\nTotal Q2: ${total_revenue_q2:,}',
        ha='center', fontsize=10, color='green')

# Show the plot
plt.show()
```
