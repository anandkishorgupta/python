```python
import pandas as pd
df=pd.read_excel("Security Questionnaire Dataset.xlsx")
df
```

```python

# Function to capitalize first letter after a period
def capitalize_after_period(sentence):
    result = ''
    capitalize_next = True

    for char in sentence:
        if capitalize_next and char.isalpha():
            result += char.upper()
            capitalize_next = False
        else:
            result += char
            if char == '.':
                capitalize_next = True

    return result

# Apply the function to the 'Question' column
df['Question'] = df['Question'].apply(capitalize_after_period)

# Display the modified DataFrame
df

```
