```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('sales.csv')

# Convert 'Sales Revenue' column to numeric (remove "$" and convert to float)
df['Sales Revenue'] = df['Sales Revenue'].replace('[\$,]', '', regex=True).astype(float)
# Group by 'Sales Channel' and calculate total revenue for each channel
channel_revenue = df.groupby('Sales Channel')['Sales Revenue'].sum().sort_values()

# Plot the bar chart
plt.bar(channel_revenue.index, channel_revenue.values, color='skyblue')
plt.title('Total Revenue by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file
df = pd.read_csv('sales.csv')

# Cleaning Sales Revenue column to convert it into numeric type
df['Sales Revenue'] = df['Sales Revenue'].str.replace('$', '').str.replace(',', '').astype(float)

# Grouping data by Sales Channel and calculating sum of Sales Revenue
grouped_data = df.groupby('Sales Channel')['Sales Revenue'].sum()

# Creating a table to display the results
table = pd.DataFrame({'Sales Channel': grouped_data.index,
                      'Sum of Sales Revenue': grouped_data.values})
# table.loc[len(table)] = ['Grand Total', table['Sum of Sales Revenue'].sum()]

# Displaying the table
table
```

```python
# Creating a bar chart
plt.bar(table['Sales Channel'], table['Sum of Sales Revenue'])

# Adding labels and title
plt.xlabel('Sales Channel')
plt.ylabel('Sum of Sales Revenue')
plt.title('Total Sales Revenue by Sales Channel')

# Displaying the chart
plt.show()
```
