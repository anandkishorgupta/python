```python
# Convert the 'Sales Revenue' and 'Advertising Spend' columns to numeric values
df['Sales Revenue'] = df['Sales Revenue'].replace('[\\$,]', '', regex=True).astype(float)
df['Advertising Spend'] = df['Advertising Spend'].replace('[\\$,]', '', regex=True).astype(float)

df.head()
```

```python
# Group by Product Category and Quarter, and calculate the total sales for each category in Q1 and Q2

sales_q1_q2 = df.groupby(['Product Category', 'Quarter'])['Sales Revenue'].sum().unstack()

# Calculate the percentage change in sales between Q1 and Q2
sales_q1_q2['Percentage Change'] = ((sales_q1_q2['Q2'] - sales_q1_q2['Q1']) / sales_q1_q2['Q1']) * 100

# Sort the table based on the percentage change in descending order
sales_q1_q2 = sales_q1_q2.sort_values(by='Percentage Change', ascending=False)

# Add meaningful column names and format the percentage change column
sales_q1_q2.columns = ['Q1 Sales', 'Q2 Sales', 'Percentage Change']
sales_q1_q2['Percentage Change'] = sales_q1_q2['Percentage Change'].map('{:.2f}%'.format)

# Display the result
sales_q1_q2
```
